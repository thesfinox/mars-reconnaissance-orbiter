## First Steps

We first need to instantiate the `HSIMars` class:

```python
hsi = HSIMars(
    hdr_path="../data/HC_frt0000580c_07_if164j_ter3.hdr",
    annotations="../data/HC_ground_truth.mat",
)
```

We then proceed with the extraction of the HSI:

```python
img = hsi.get_img()
type(img)
```

The object is a `namedtuple` containing several attributes, which can be accessed using the usual Python interface `img.<attr>`:

- `hsi`: the HSI data,
- `wavelength`: the list of wavelengths (units: nm),
- `shape`: the shape of the image,
- `height`: the height of the image,
- `width`: the width of the image,
- `channels`: the number of channels in the image,
- `dtype`: the data type of the HSI.

```python
print("img.hsi -->", type(img.hsi))
print("img.wavelength -->", type(img.wavelength))
print("img.shape -->", type(img.shape))
print("img.height -->", type(img.height))
print("img.width -->", type(img.width))
print("img.channels -->", type(img.channels))
print("img.dtype -->", type(img.dtype))
```

We can also recover the annotations using the same principle:

```python
ann = hsi.get_annotations()
type(ann)
```

The output is another `namedtuple`, containing the following attributes:

- `labels`: the annotation data,
- `shape`: the shape of the image,
- `height`: the height of the image,
- `width`: the width of the image,
- `dtype`: the data type of the annotation.

```python
print("ann.labels -->", type(ann.labels))
print("ann.shape -->", type(ann.shape))
print("ann.height -->", type(ann.height))
print("ann.width -->", type(ann.width))
print("ann.dtype -->", type(ann.dtype))
```

If you prefer, you can also call both of these outputs at the same time:

```python
img, ann = hsi.data()
type(img), type(ann)
```

## Visualisation

You can also visualise the annotations and the image (in **fake colours**) using `opencv` (a new window will open, close it with "q"):

```python
hsi.display_hsi()
```

You can visualise the annotations (close the new window with "q"):

```python
hsi.display_annotations()
```

Or you can display all of the information at once (close again with "q"):

```python
hsi.display()
```

We can also plot a particular spectrum:

```python
hsi.plot_spectra((128, 145))
```

```python
hsi.plot_spectra([[128, 145], [129, 146], [128, 146], [129, 145]], bands=True)
```

```python
hsi.plot_spectra(
    [[128, 145], [129, 146], [128, 146], [129, 145]],
    convex_hull=True,
    bands=True,
)
```

Or even the histogram of a particular band:

```python
hsi.plot_histogram(band=146)
```

```python
hsi.plot_histogram(band=1493.0)
```
